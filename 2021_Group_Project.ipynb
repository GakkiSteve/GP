##### Exploration

As a team, you need to identify _good_ set of data sources. A _good_ datasource for this projects is a source of data that:  
1. you can read and handle, or learn to read and handle in a very short time. Formats that are too exotique may be very interesting, but they may also require a too long time to import into anything familiar to you. To be fair, most of the datasets you can find in the sources we suggest you are in a good format. _pro tip_: avoid data presented in pdf tables, unless you are ready to do a lot of hand work (if you do, it's great).
2. it is not already in the most suitable data model. If the source dataset is already perfect, there's little you can do and you won't have the opportunity to exhibit your skills. Government or NGO data is usually quite dirty, so they represent good choices.
3. it is _interesting_ for you. I don't have a strong interest in cars, and for me working a month on car data set would be a boring hell. But maybe you LOVE cars, than that dataset would be perfect for you. Try to consider your passions, or look at the news to see what may be a hot topic. Avoid abused datasets (i.e., _iris_ or _mtcars_ or _titanic_ or ...): we have already seen all possible projects about that stuff.
4. it is _joined_ or _joinable_ to other data sources. Consider the presence of common identifiers (things as: region names, product ids, years or dates, ...) that you may use to connect two or more data sources and tell a story (great data stories often emerge from connections).

## wrangling life expectancy df
All our work is based on the combination of life expectancy and other datasets. We can download the life expectancy dataset as a csv file from the website 'http://hdr.undp.org/en/data', but after reading it, we find that it cannot be used directly.

```r
life_expectancy <- "Life expectancy at birth (years).csv" %>%
  read_csv(col_types = cols(
      HDI = col_double() # read HDI as double, if you read it directly, it will be chr type.
      )
  )
```

```r
life_expectancy %>%
head(5)
```

```r
# Calculate the mean of life expectancy from 1900 to 2019 using for loop.

for(i in 1:191) {
    now_life_df <- life_expectancy[i,]
    now_life_df$mean_life <- rowMeans(now_life_df[3:32])
    if(i == 1) {
        life_df <- now_life_df
        } else{
        life_df <- rbind(life_df, now_life_df)
    }

}
```

```r
# Sort life_df by Life_Rank

life_df <- life_df %>%
    arrange(desc(mean_life)) %>% 
    na.omit() # Omit missing values
```

```r
life_df %>%
head(5)
```

```r
write.csv(life_df,"life_df.csv" ) # save life_df as csv file in local, then we can read it more conveniently
```

```r
# Create function to convert to long format

convert_to_long <- function(data) {
    gather(data, Year, Factor, -Life_Rank, -Country) %>%
    transform(Year = as.numeric(Year), Factor = as.numeric(Factor), check.names=FALSE) %>%
    na.omit() 
}
```

In this way, we get countries whose life expectancy ranges from long to short

After processing this dataset into a format that we are satisfied with, we can start the formal work

## Economic aspect

The first part is about economy and it includes GDP per captia and unemployment rate of each country.

### GDP per captia

Read in the information from the CSV file GDP per capita (2017 PPP $).csv

CSV file from the undp website, Source: http://hdr.undp.org/en/indicators/194906

```r
GDP_df <- read_csv('GDP per capita (2017 PPP $).csv')
GDP_df <- inner_join(life_df,GDP_df, by='Country') # Combine life_df and GDP_df by Country
GDP_df <- subset(GDP_df, select = -c(HDI)) # Don't need HDI now
```

```r
glimpse(GDP_df) # By using glismpse(), we can know there is 188 rows of this dataframe
```

```r
# Create vectors for the Life_Rank in each group
highest <- c(1:10) # The country with the longest lifespan (1-10)
high <- c(42:51) # 75 Percentile (42-51)
middle <- c(91:100) # 50 Percentile (91-100)
low <- c(138:147) # 25 Percentile (138-147)
lowest <-c(179:188) # The country with the shortest lifespans (179-188)
```

```r
# Create dataframe for each part of GDP_df

life_time_highest <- subset(GDP_df, Life_Rank %in% highest)
life_time_highest$Life_Rank <- '1_highest'

life_time_high <- subset(GDP_df, Life_Rank %in% high)
life_time_high$Life_Rank <- '2_high'

life_time_middle <- subset(GDP_df, Life_Rank %in% middle)
life_time_middle$Life_Rank <- '3_middle'

life_time_low <- subset(GDP_df, Life_Rank %in% low)
life_time_low$Life_Rank <- '4_low'

life_time_lowest <- subset(GDP_df, Life_Rank %in% lowest)
life_time_lowest$Life_Rank <- '5_lowest'
```

```r
# Bind all dataframe into new dataframe
life_time_df <- rbind(life_time_highest, life_time_high, life_time_middle, life_time_low, life_time_lowest) 
```

```r
# Convert dataframe to long type
life_time_df <- convert_to_long(life_time_df) %>% rename(GDP_ppp = Factor)
```

```r
glimpse(life_time_df) # We can see all year value are in Year column now
```

```r
ggplot() +
    # Plot the GDP per captia in comparison to life expectancy
    geom_line(data = life_time_df, aes(x=Year, y=GDP_ppp, group=Country, color=Life_Rank)) +
    scale_color_manual(values = c('dodgerblue', 'olivedrab', 'gold', 'tomato','blue')) + theme_bw() + 
    ggtitle('GDP in comparison to Life expectancy') + xlab('Year') + ylab('GDP_ppp')

ggsave("GDP per captia.png")
```

```r
life_time_df %>%
group_by(Country) %>%
    ggplot(mapping = aes(x = Year, y = GDP_ppp, color = Life_Rank))+ geom_point(alpha = 0.6, size = 3)+
    geom_line(data = life_time_df, aes(x=Year, y=GDP_ppp, group=Country, color=Life_Rank), size = 2) +
    ggtitle('GDP in comparison to Life expectancy') + xlab('Year') + ylab('GDP_ppp')

ggsave("GDP per captia 2.png")
```

Maybe it’s because the lines in this picture are too scattered, and it’s hard to show any clear conclusions.

We need to use the mean value of each class to make a clearer picture

```r
GDPgrouped <- life_time_df %>% 
        group_by(Life_Rank, Year) %>% 
        summarise(GDP_ppp = mean(GDP_ppp))
```

```r
ggplot(life_time_df, aes(x = Year, y = GDP_ppp, color = Life_Rank)) +
    geom_line(aes(group = Country), alpha = .3) + 
    geom_line(data = GDPgrouped, alpha = .8, size = 3) + 
    theme_bw() + ggtitle('Average GDP for each Life_Rank') +
    xlab('Year') + ylab('GDP_ppp')

ggsave("Average GDP per captia.png")
```

### Pollution

All data is scarped from website https://www.numbeo.com/pollution.

```r
link_for_pollution = "https://www.numbeo.com/pollution/rankings_by_country.jsp?title=2020"
page_for_pollution <- link_for_pollution %>% read_html()
```

```r
# Write a function to get data of different year.
get_one_year_p <- function(n){
    link_for_year <- glue("https://www.numbeo.com/pollution/rankings_by_country.jsp?title={n}")
    page_of_year <- link_for_year %>% read_html()
       #get html into R
    
    #create dataframe,similar to what we do above
    node = "#t2"
    table = page_of_year %>%
      html_nodes(node)
    pollution_df = as.data.frame(table %>% html_table(fill = TRUE))
    pollution_df = pollution_df[,-c(1,4)]
    pollution_df$year <- n
    return(pollution_df)
}
```

```r
p_total = get_one_year_p(2014) # try to scrape
```

```r
glimpse(p_total) # There are 115 rows, so we need to redistribute the range of each category
```

```r
# Create new vectors for the Life_Rank in each group
highest <- c(1:10) # The country with the longest lifespan (1-10)
high <- c(23:32) # 75 Percentile (23-32)
middle <- c(52:61) # 50 Percentile (52-61)
low <- c(81:90) # 25 Percentile (81-90)
lowest <-c(106:115) # The country with the shortest lifespans (106-115)
```

```r
# Collect all years of pollution that can be collected
for(n in 2015:2021)
    {
    tmp = get_one_year_p(n)
    p_total = rbind(p_total,tmp)
}
```

```r
#pollution index for each country fro 2014-2021
pollution_df <- inner_join(life_df,p_total, by='Country') # Combine life_df and pollution_df by Country
```
